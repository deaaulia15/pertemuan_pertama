```python
#import library
from bs4 import BeautifulSoup
import requests

# pandas adalah library open source pada pyton yang sering digunakan
# untuk memproses data yang meliputi pembersihan data,manipulasi data,
# hingga melakukan analisis data
import pandas as pd
```

```python
# List kosong untuk menampung hasil dari scrapping data
data1 = []

# Membuat perulangan secara otomatis untuk mengambil judul artikel dan link dari halaman 1 - halaman 50
for page in range(1, 2):
    if page == 1:
        url = "https://turnbackhoax.id/?s=aceh"
    else:
        url = "https://turnbackhoax.id/page/" + str(page) +"/?s=aceh"

    # Request URL
    req = requests.get(url)

    # Merubah format dari yang semula HTML menjadi BeautifulSoup
    soup = BeautifulSoup(req.text, "html.parser")

    # Memanggil seluruh artikel yang berada di halaman luar
    column = soup.findAll("div", "mh-loop-content mh-clearfix")

    for scraping in column:
        text = scraping.find("h3", "entry-title mh-loop-title").text.strip().split("\n")
        link = scraping.find("a", {"rel":"bookmark"})["href"]
        print(text, link)

        data1.append([text, link]) # Memasukkan data scraping ke dalam list data
```

```python
# Memasukkan hasil scrapping ke dalam DataFrame
df1 = pd.DataFrame(data1)
df1.head()
```

```python
#Mengganti nama kolom pada DataFrame
df1.rename(columns={0:"BERITA",1:"URL"}, inplace= True)
df1.head()
```

```python
df1.info()
```

```python
# List kosong untuk menampung hasil dari scrapping data
data2 = []

# Membuat perulangan secara otomatis untuk menggambil judul artikel dan link dari halaman 1 - halaman 50
for page in range(1, 2):
    if page ==  1:
      url = "https://turnbackhoax.id/?s=tsunami"
    else:
      url = "http://turnbackhoax.id/page/" + str(page) +"/?s+tsunami"

    #Request URL
    req = requests.get(url)

    # Merubah format dari yang semula HTML menjadi BeautifulSoup
    soup = BeautifulSoup(req.text, "html.parser")

    # Memanggil seluruh artikel yang berada di halaman luar
    column = soup.findAll("div", "mh-loop-content mh-clearfix")

    for scraping in column:
        text = scraping.find("h3", "entry-title mh-loop-title").text.strip().split("\n")
        link = scraping.find("a", {"rel":"bookmark"})["href"]
        print(text, link)

        data2.append([text, link]) # Masukkan data scraping ke dalam list data
```

```python
# Memasukkan hasil scrapping ke dalam DataFrame
df2 = pd.DataFrame(data2)
df2.head()
```

```python
# Mengganti nama kolom pada DataFrame
df2.rename(columns={0:"BERITA", 1:"URL"}, inplace= True)
df2.head()
```

```python
# Membuat DataFrame beru untuk menampung kedua DataFrame hasil scraping
data_scrap = [df1, df2]

# Menggabungkan DataFrame
data_scraping = pd.concat(data_scrap)

# Head adalah akan menampilkan 5 baris data teratas
data_scraping.head()
```

```python
# info digunakan untuk menampilkan informasi detail tentang dataFrame,
# seperti jumlah baris data, nama-nama kolom berserta
# jumah data dan tipe datanya, dan sebagainya
data_scraping.info()
```

```python
# Tail adalah akan menampilkan 5 baris data terbawah
data_scraping.tail()
```

```python
from sqlalchemy.sql.expression import false
# Menjadikan DataFrame ke dalam.csv
data_scraping.to_csv("Data scraping.csv",index= false)
```
